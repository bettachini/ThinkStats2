## Ensayos de leer formato de stata

```python
def creaDataType(file = './2002FemPreg.dct'):
    '''
    Debe leer *.dct de Stata y generara el datatype de numpy
    Para esto linea por linea
        tomar la denominación de la columna 3 -> nombre (serie pandas)
        y tomar el tipo de dato de la columna 4 -> tipo (serie pandas)
    '''
    test = pd.read_csv(file, delim_whitespace=True)[:-1]
    nombre = list(test['infile'])
    tipo = list(test['dictionary'])
    for i in range(len(tipo)):
        if tipo[i]=='%12s':
            tipo[i] = 'U12'
        elif tipo[i]=='%1f':
            tipo[i] = 'b1'
        elif tipo[i]=='%2f':
            tipo[i] = 'b2'
        elif tipo[i]=='%3f':
            tipo[i] = 'b3'
        elif tipo[i]=='%4f':
            tipo[i] = 'b4'
        elif tipo[i]=='%18f':
            tipo[i] = 'b18'
    cas = zip(nombre,tipo)
    cus= []
    for i in cas:
        cus.append(i)
    dtyp = np.dtype(cus)
    return dtyp

clap = creaDataType()
clap
```
